```python
# Converting json to csv
def json2csv(json_name, csv_name, key, subKey):
    with open(json_name) as json_file:
        data = json.load(json_file)
    transaction_data = data[key][subKey]
    data_file = open(csv_name, 'w') 
    csv_writer = csv.writer(data_file)
    count = 0
    for tran in transaction_data:
        if count == 0: 
            # Writing headers of CSV file 
            header = tran.keys()
            csv_writer.writerow(header) 
            count += 1
        # Writing data of CSV file 
        csv_writer.writerow(tran.values())
    data_file.close() 
```

```python
# This function is used to identfiy the life stage of person using "age"
def identifyLifeStage(age):
    if age < 20:
        return "Teenager"
    elif age < 22:
        return "Student"
    elif age < 30:
        return "Early employeed"
    elif age < 60:
        return "Mature employeed"
    else: 
        return "Retired"

identifyLifeStage(profileDataDict['age'])
```

```python
# Budget split of the person

income = demographicsDict["income"]
# Percent wise split of categories  (+5 for upper range)
budgetSplitPercent = {
    'Insurance': 5,
    'Health Care': 5,
    'Food': 10,
    'Transportation': 10,
    'Personal': 5,
    'Recreation': 5,
    'Utilities': 5,
    'EMI/Housing': 25,
    'Savings': 10,
}

lowLimitBudget = {
    'Insurance': 0,
    'Health Care': 0,
    'Food': 0,
    'Transportation': 0,
    'Personal': 0,
    'Recreation': 0,
    'Utilities': 0,
    'EMI/Housing': 0,
    'Savings': 0
} 

upperLimitBudget = {
    'Insurance': 0,
    'Health Care': 0,
    'Food': 0,
    'Transportation': 0,
    'Personal': 0,
    'Recreation': 0,
    'Utilities': 0,
    'EMI/Housing': 0,
    'Savings': 0
} 

for category in budgetSplitPercent:
    lowLimitBudget[category] = budgetSplitPercent[category] * .01 * income
    upperLimitBudget[category] = (budgetSplitPercent[category] + 5) * .01 * income

print("Low:", lowLimitBudget)
print("upper:", upperLimitBudget)
```

```python
# Evaluating spends as per budget

categoryToKeyMapper = {
    'Insurance': ['Insurance Premium'],
    'Health Care': ['Health Care'],
    'Food': ['Food'],
    'Transportation': ['Transportation'],
    'Personal': ['Personal Care', 'Shopping', 'Transfer to Wallet', 'Purchase by Card'],
    'Recreation': ['Entertainment'],
    'Utilities': ['Utilities'],
    'EMI/Housing': ['EMIs'],
    'Savings': ['']
}

budgetValues = {
    'Insurance': 0,
    'Health Care': 0,
    'Food': 0,
    'Transportation': 0,
    'Personal': 0,
    'Recreation': 0,
    'Utilities': 0,
    'EMI/Housing': 0,
    'Savings': 0
}

budgetEval = {
    'Insurance': '',
    'Health Care': '',
    'Food': '',
    'Transportation': '',
    'Personal': '',
    'Recreation': '',
    'Utilities': '',
    'EMI/Housing': '',
    'Savings': ''
}

totalAmountExcludingSavings = 0

for category in budgetEval:
    catTotal = 0
    for catKey in categoryToKeyMapper[category]:
        if catKey != '':
            catTotal += categoriesDict[catKey]["totalDebitAmount"] - categoriesDict[catKey]["totalCreditAmount"]
        else: # For savings
            catTotal = demographicsDict["income"] - totalAmountExcludingSavings - categoriesDict["Merchant Expenditure"]["totalDebitAmount"]
    budgetValues[category] = catTotal
    if catTotal < lowLimitBudget[category]:
        budgetEval[category] = 'below'
    elif catTotal >= lowLimitBudget[category] and catTotal <= upperLimitBudget[category]:
        budgetEval[category] = 'on-par'
    else:
        budgetEval[category] = 'above'  
    totalAmountExcludingSavings += catTotal

print(budgetValues)

budgetEval
```

```python
# Different feature flags

featureFlags = {
    'investor': False,
    'mutualFunds': False,
    'FDs': False,
    'insured': False,
    'loans': False,
    'heavyShopper': False,
    'heavyOutsideFood': False,
    'heavyEntertainment': False,
    'heavyUtilities': False,
    'heavyTransportation': False,
    'heavyHealthCare': False,
    "heavyInvestor": False,
    "heavySaver": False,
    "heavyDebit": False,
    "heavyCredit": False,
    "creditCardUser": False,
    'paymentMissedRecently': True,
    'lumpSumReceivedRecently': True
}

def evaluateFlags():
    
    if investmentDataDict["totalDebitAmount"] and investmentDataDict["totalDebitAmount"] > 0:
        featureFlags['investor'] = True
    if mutualFundsObject["amount"] and float(mutualFundsObject["amount"]) > 0:
        featureFlags["mutualFunds"] = True
    if investmentDataDict["subCategoryAnalysis"] and investmentDataDict["subCategoryAnalysis"]["Fixed/Recurring Deposits"]["totalCreditAmount"] > 0:
        featureFlags["FDs"] = True
    if categoriesDict["Insurance Premium"] and categoriesDict["Insurance Premium"]["totalDebitAmount"] > 0:
        featureFlags["insured"] = True
    if categoriesDict["EMIs"] and categoriesDict["EMIs"]["totalDebitAmount"] > 0:
        featureFlags["loans"] = True
    if budgetEval['Personal'] and (budgetEval['Personal'] == 'above' or budgetEval['Personal'] == 'on-par') :
        featureFlags['heavyShopper'] = True
    if budgetEval['Food'] and (budgetEval['Food'] == 'above' or budgetEval['Food'] == 'on-par'):
        featureFlags["heavyOutsideFood"] = True
    if budgetEval["Recreation"] and (budgetEval["Recreation"] == 'above' or budgetEval["Recreation"] == 'on-par'):
        featureFlags["heavyEntertainment"] = True
    if budgetEval["Utilities"] and (budgetEval["Utilities"] == 'above' or budgetEval['Utilities'] == 'on-par'):
        featureFlags["heavyUtilities"] = True
    if budgetEval['Transportation'] and (budgetEval['Transportation'] == 'above' or budgetEval['Transportation'] == 'on-par'):
        featureFlags["heavyTransportation"] = True
    if budgetEval["Health Care"] and (budgetEval["Health Care"] == 'above' or budgetEval['Health Care'] == 'on-par'):
        featureFlags["heavyHealthCare"] = True
    if investmentDataDict["totalDebitAmount"] and investmentDataDict["totalDebitAmount"] > demographicsDict["income"] * 10 * .01:
        featureFlags["heavyInvestor"] = True
    if budgetEval["Savings"] and (budgetEval["Savings"] == 'above'):
        featureFlags["heavySaver"] = True
    if keySummary["creditDebitRatio"] and keySummary["creditDebitRatio"] <= 1:
        featureFlags["heavyDebit"] = True    
    if keySummary["creditDebitRatio"] and keySummary["creditDebitRatio"] > 1:
        featureFlags["heavyCredit"] = True
    if categoriesDict['Credit Card Payment'] and categoriesDict['Credit Card Payment']["totalDebitAmount"] > 0:
        featureFlags["creditCardUser"] = True
    
evaluateFlags()

featureFlags
```

```python
# Dividing each individual into different finance personality types-> One can fit into multiple as well

financePersonalityTypes = ["investors", "savers", "big spenders", "debtors", "shoppers"]

# Atleast 2/3rd of the features should be satisfied for each personality type
personalityFeatures = {
    "investors": ["heavyInvestor"],
    "savers": ["heavySaver"],
    "big spenders": ["heavyCredit", "heavyShopper", "heavyEntertainment", "heavyOutsideFood", "heavyTransportation"],
    "debtors": ["heavyDebit"],
    "shoppers": ["heavyShopper"]    
}


personalityIdentifier = {
    "investors": False,
    "savers": False,
    "big spenders": False,
    "debtors": False,
    "shoppers": False    
}


for personalityType in personalityFeatures:
    
    totalFeatures = len(personalityFeatures[personalityType])
    if totalFeatures == 1:
        requiredNoOfFeaturesToSatisfy = 1
    else:    
        requiredNoOfFeaturesToSatisfy = int(2 * totalFeatures/3)
    trueCount = 0

    for feature in personalityFeatures[personalityType]:
        if featureFlags[feature]: 
            trueCount += 1
    if trueCount >= requiredNoOfFeaturesToSatisfy:
        personalityIdentifier[personalityType] = True

personalityIdentifier
```

```python
# Putting the user into miscellaneous segments

miscTypes = {
    "paymentForgetful": False,
    "lumpSumReceiver": False,
    "underCoveredInsurance": False, # Cover should be atleast 10x the annual income
    "studentLoanRequired": False, # Based on life stage
    "foodie": False, # Based on heavyOutsideFood,
    "shopper": False, #  "   "  heavyShopping
    "entertainMax": False, # "  " heavyEntertainment
    "creditCardUser": False, # creditCardUser feature flag
    "nomineeMF": False, # Based on mutualFundsObject 'nominee'
}

annualIncome = (categoriesDict["Salary"]["totalCreditAmount"]/profileDataDict['monthsOfData']) * 12 # To find aprx annual income
# print(annualIncome, float(insuranceSummaryObj['sumAssured']))

def evaluateMiscTypes():
    if featureFlags['paymentMissedRecently']:
        miscTypes["paymentForgetful"] = True
    if featureFlags['lumpSumReceivedRecently']:
        miscTypes['lumpSumReceiver'] = True
    if float(insuranceSummaryObj['sumAssured']) < annualIncome * 10:
        miscTypes['underCoveredInsurance'] = True
    if demographicsDict['lifeStage'] == 'Student':
        miscTypes["studentLoanRequired"] = True
    if featureFlags['heavyOutsideFood']:
        miscTypes["foodie"] = True
    if featureFlags['heavyShopper']:
        miscTypes["shopper"] = True
    if featureFlags["heavyEntertainment"]:
        miscTypes["entertainMax"] = True
    if featureFlags["creditCardUser"]:
        miscTypes["creditCardUser"] = True
    if mutualFundsObject["nominee"] == 'REGISTERED':
        miscTypes['nomineeMF'] = True

evaluateMiscTypes()
miscTypes
```
